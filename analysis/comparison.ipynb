```python
models = [
    "KNN", 
    "Random Forest", 
    "SVM (Our)", 
    "SVM (Linear)", 
    "SVM (RBF)", 
    "SVM (Polynomial)", 
    "Logistic Regression (Our)", 
    "CV SVM"
]
train_accuracy = [0.8791, 1.0000, 0.7944, 0.7955, 0.8771, 0.7727, 0.6488, 0.7955]
test_accuracy = [0.8347, 0.9050, 0.8678, 0.867769, 0.876033, 0.760331, 0.6529, 0.8678]
precision = [0.8347, 0.8898, 0.846154, 0.857143, 0.853846, 0.681818, 0.6044, 0.857143]
recall = [0.8347, 0.9262, 0.901639, 0.885246, 0.909836, 0.983607, 0.9016, 0.885246]
f1_score = [0.8347, 0.9076, 0.873016, 0.870968, 0.880952, 0.805369, 0.7237, 0.870968]
auroc = [0.8347, 0.9048, 0.867486, 0.867623, 0.875751, 0.758470, 0.6508, 0.867623]
runtime = [1.1612, 0.16, 0.7686, 0.0411, 0.0475, 0.0249, 3.8238, 15.0336]

# Visualization for performance metrics
metrics = ["Train Accuracy", "Test Accuracy", "Precision", "Recall", "F1 Score", "AUROC"]
metric_values = [train_accuracy, test_accuracy, precision, recall, f1_score, auroc]

# Plotting each metric
for i, metric in enumerate(metrics):
    plt.figure(figsize=(12, 6))
    plt.bar(models, metric_values[i], color='skyblue', alpha=0.9)
    plt.title(f"{metric} Comparison Across Models", fontsize=16)
    plt.ylabel(metric, fontsize=14)
    plt.xlabel("Models", fontsize=14)
    plt.xticks(rotation=45, fontsize=12)
    plt.tight_layout()
    plt.show()

# Visualization for runtime
plt.figure(figsize=(12, 6))
plt.bar(models, runtime, color='purple', alpha=0.7)
plt.title("Runtime Comparison Across Models", fontsize=16)
plt.ylabel("Runtime (s)", fontsize=14)
plt.xlabel("Models", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.tight_layout()
plt.show()
```

```python
metrics = ["Train Accuracy", "Test Accuracy", "Precision", "Recall", "F1 Score", "AUROC"]
metric_values = [train_accuracy, test_accuracy, precision, recall, f1_score, auroc]
colors = ['blue', 'green', 'red', 'purple', 'orange', 'cyan']

# Create a combined bar chart
x = range(len(models))  # x positions for models
width = 0.12  # width of each bar

plt.figure(figsize=(14, 8))
for i, (metric, values) in enumerate(zip(metrics, metric_values)):
    plt.bar(
        [pos + i * width for pos in x],  # shift bars for each metric
        values, 
        width, 
        label=metric, 
        color=colors[i],
        alpha=0.8
    )

# Add labels and legend
plt.title("Performance Metrics Comparison Across Models", fontsize=16)
plt.ylabel("Scores", fontsize=14)
plt.xlabel("Models", fontsize=14)
plt.xticks([pos + (len(metrics) - 1) * width / 2 for pos in x], models, rotation=45, fontsize=12)
plt.legend(title="Metrics", fontsize=12)
plt.tight_layout()

# Show plot
plt.show()
```
